# In-class Variational Method in One Basis Function

We will encounter the following two integrals frequently for this problem

$\int_{-\infty}^{\infty} e^{-2\alpha (x-x_0)^2} dx = \sqrt{\frac{\pi}{2\alpha}}$

and 

$\int_{-\infty}^{\infty} x^2e^{-2\alpha x^2} dx = \frac{1}{2}\sqrt{\frac{\pi}{(2\alpha)^3}} = \frac{1}{4\alpha} \sqrt{\frac{\pi}{2\alpha}}$.

Notice that the second of these has the exponent and the prefactor centered around $x=0$.  Thus is we get an integral of the form

$\int_{-\infty}^{\infty} x^2e^{-2\alpha (x-x_0)^2} dx$

we must rearrange it such that the prefactor and the exponent have the same center.  This can be done by either integration by parts or substitution.

## Variational Method for One Basis Function (due by 5pm Monday 10/8/18)

1. What is the energy expectation value for the trial wavefunction, $\phi(x) = c_0 e^{-\alpha (x-x_0)^2}$, for a 1D harmonic oscillator ($\hat{H} = \hat{K} + \hat{V} = -\frac{\hbar^2}{2\mu}\frac{d^2}{dx^2} + \frac{\mu\omega^2}{2}x^2$ :note that I have used $k = \mu\omega^2$)? 

 1. The goal is to derive an expression for $\langle E \rangle_\phi$. Below I breakdown the problem and show the solutions for each component so that you can check your work.

    a. $\langle E \rangle_\phi = \frac{\langle \phi|\hat{H}|\phi\rangle}{\langle \phi | \phi \rangle} = \frac{\langle \phi|\hat{K}|\phi\rangle + \langle \phi|\hat{V}|\phi\rangle}{\langle \phi | \phi \rangle}$

    b. $\langle \phi| \phi \rangle = \int_{-\infty}^{\infty} c_0 e^{-\alpha (x-x_0)^2} c_0 e^{-\alpha (x-x_0)^2} dx = c_0^2\sqrt{\frac{\pi}{2\alpha}}$
    
    **This one was relatively straight forward**
    
    $\int_{-\infty}^{\infty} c_0 e^{-\alpha (x-x_0)^2} c_0 e^{-\alpha (x-x_0)^2} dx \\
    = c_0^2 \int_{-\infty}^{\infty} e^{-2\alpha (x-x_0)^2} dx \\
    = c_0^2 \sqrt{\frac{\pi}{2\alpha}}$

    c. $\langle \phi|\hat{K}|\phi\rangle = \int_{-\infty}^{\infty} c_0 e^{-\alpha (x-x_0)^2}\left[ -\frac{\hbar^2}{2\mu}\frac{d^2}{dx^2} \right]c_0 e^{-\alpha (x-x_0)^2} dx = \frac{c_0^2\hbar^2\alpha}{2\mu}\sqrt{\frac{\pi}{2\alpha}} $
    
    **For this one we should start by considering what the second derivative operator acting on the basis function is**
    
    $ \frac{d^2}{dx^2}c_0 e^{-\alpha (x-x_0)^2}=-2c_0\alpha\frac{d}{dx}\left[(x-x_0)e^{-\alpha (x-x_0)^2}\right]
    \\= -2c_0\alpha \left[ e^{-\alpha (x-x_0)^2} -2\alpha(x-x_0)^2e^{-\alpha (x-x_0)^2}\right]$
    
    **Notice that you have to perform the product rule of differentiation on $\frac{d}{dx}\left[(x-x_0)e^{-\alpha (x-x_0)^2}\right]$.  If we now multiply this by $-\frac{\hbar^2}{2\mu}$ we get**
    
    $-\frac{\hbar^2}{2\mu}\frac{d^2}{dx^2}c_0e^{-\alpha (x-x_0)^2}=-\frac{\hbar^2}{2\mu}\left(-2c_0\alpha \left[ e^{-\alpha (x-x_0)^2} -2\alpha(x-x_0)^2e^{-\alpha (x-x_0)^2}\right]\right) \\
    = \frac{\hbar^2c_0\alpha}{\mu}\left[ e^{-\alpha (x-x_0)^2} -2\alpha(x-x_0)^2e^{-\alpha (x-x_0)^2}\right] $
    
    **Now plug back into equation:**
    
    $\langle \phi|\hat{K}|\phi\rangle = \frac{\hbar^2c_0^2\alpha}{\mu}\int_{-\infty}^{\infty} e^{-\alpha (x-x_0)^2} \left[ e^{-\alpha (x-x_0)^2} -2\alpha(x-x_0)^2e^{-\alpha (x-x_0)^2} \right] dx$ 
    
    $= \frac{\hbar^2c_0^2\alpha}{\mu} \left[ \int_{-\infty}^{\infty} e^{-2\alpha (x-x_0)^2} dx -2\alpha \int_{-\infty}^{\infty} (x-x_0)^2e^{-\alpha (x-x_0)^2} dx\right] \\
    =\frac{\hbar^2c_0^2\alpha}{\mu} \left[  \sqrt{\frac{\pi}{2\alpha}} - 2\alpha \frac{1}{2}\sqrt{\frac{\pi}{(2\alpha)^3}}\right] \\
     = \frac{\hbar^2c_0^2\alpha}{\mu}\sqrt{\frac{\pi}{2\alpha}}\left[1-\frac{1}{2}\right] 
     \\ = \frac{\hbar^2c_0^2\alpha}{2\mu}\sqrt{\frac{\pi}{2\alpha}}$
    
    d. $ \langle \phi|\hat{V}|\phi\rangle = \int_{-\infty}^{\infty} c_0 e^{-\alpha (x-x_0)^2}\left[ \frac{\mu\omega^2}{2}x^2 \right]c_0 e^{-\alpha (x-x_0)^2} dx = \frac{c_0^2\mu\omega^2}{2}\sqrt{\frac{\pi}{2\alpha}}\left[\frac{1}{4\alpha}+x_0^2\right] $

    **The first step is somewhat straightforward algebraic manipulation to yield**
    
    $ \langle \phi|\hat{V}|\phi\rangle = \frac{\mu\omega^2c_0^2}{2}\int_{-\infty}^{\infty}  x^2e^{-2\alpha (x-x_0)^2}dx $
    
    **To simplify you can either do integration by parts or substitution.  Here I do substitution with $u=(x-x_0)$ which rearranges to give $x=u+x_0$ and $dx=du$.  Performing this yields**
    
    $ \langle \phi|\hat{V}|\phi\rangle = \frac{\mu\omega^2c_0^2}{2}\int_{-\infty}^{\infty}  (u+x_0)^2e^{-2\alpha u^2}du $
    
    **Expand $(u+x_0)^2 = u^2+2ux_0+x_0^2$ to get**
    
    $ \langle \phi|\hat{V}|\phi\rangle = \frac{\mu\omega^2c_0^2}{2}\left[\int_{-\infty}^{\infty} u^2e^{-2\alpha u^2}du + 2x_0\int_{-\infty}^{\infty} ue^{-2\alpha u^2}du + x_0^2\int_{-\infty}^{\infty} e^{-2\alpha u^2}du \right]$
    
    **Notice the middle of the three integrals on the right-hand side of the above expression is zero because it is an odd function integrated from $-\infty$ to $\infty$.  Thus we get**
    
    $ \langle \phi|\hat{V}|\phi\rangle = \frac{\mu\omega^2c_0^2}{2}\left[\frac{1}{2}\sqrt{\frac{\pi}{(2\alpha)^3}} +  x_0^2\sqrt{\frac{\pi}{2\alpha}} \right]\\
    =\frac{\mu\omega^2c_0^2}{2}\sqrt{\frac{\pi}{2\alpha}}\left[ \frac{1}{4\alpha} + x_0^2 \right] $

    e. $\langle E \rangle_\phi = \frac{\hbar^2\alpha}{2\mu} + \frac{\mu\omega^2}{2}\left[\frac{1}{4\alpha} + x_0^2 \right]$
    
    **This is simple algebraic manipulation of the three expressions derived above.**
    
    $\langle E \rangle_\phi = \frac{\langle \phi|\hat{K}|\phi\rangle + \langle \phi|\hat{V}|\phi\rangle}{\langle \phi|\phi\rangle} \\
    = \frac{\frac{\hbar^2c_0^2\alpha}{2\mu}\sqrt{\frac{\pi}{2\alpha}}+ \frac{\mu\omega^2c_0^2}{2}\sqrt{\frac{\pi}{2\alpha}}\left[ \frac{1}{4\alpha} + x_0^2 \right]}{c_0^2 \sqrt{\frac{\pi}{2\alpha}}} \\
    = \frac{c_0^2\sqrt{\frac{\pi}{2\alpha}}\left[\frac{\hbar^2\alpha}{2\mu}+ \frac{\mu\omega^2}{2}\left( \frac{1}{4\alpha} + x_0^2 \right)\right]}{c_0^2 \sqrt{\frac{\pi}{2\alpha}}} \\
    \frac{\hbar^2\alpha}{2\mu} + \frac{\mu\omega^2}{2}\left[\frac{1}{4\alpha} + x_0^2 \right]$

2. Minimize $\langle E \rangle_\phi$ with resepct to $\alpha$ and show that this yields $\alpha=\frac{\mu\omega}{2\hbar}$.  Note that this is applying the variational principle to our trial wavefunction (to non-linear coefficient $\alpha$).  Also recognize that this is dictating the width of the gaussian basis function.

   $\frac{d\langle E \rangle_\phi}{d\alpha} = \frac{d}{d\alpha} \left(\frac{\hbar^2\alpha}{2\mu} + \frac{\mu\omega^2}{2}\left[\frac{1}{4\alpha} + x_0^2 \right] \right)\\
   = \frac{\hbar^2}{2\mu} - \frac{\mu\omega^2}{2}\frac{1}{4\alpha^2} \\
   = \frac{\hbar^2}{2\mu} - \frac{\mu\omega^2}{8\alpha^2} $
   
   **Minimize by setting $\frac{d\langle E \rangle_\phi}{d\alpha} = 0$.  Then solve for $\alpha$ to find an expression for $\alpha$ that minimizes the energy**
   
   $\frac{d\langle E \rangle_\phi}{d\alpha} = 0 = \frac{\hbar^2}{2\mu} - \frac{\mu\omega^2}{8\alpha^2} \\
   \Rightarrow \frac{\hbar^2}{2\mu} = \frac{\mu\omega^2}{8\alpha^2} \\
   \Rightarrow 8\hbar^2\alpha^2 = 2\mu^2\omega^2 \\
   \Rightarrow \alpha^2 = \frac{\mu^2\omega^2}{4\hbar^2} \\
   \Rightarrow \alpha = \frac{\mu\omega}{2\hbar}$
   
3.  Minimize $\langle E \rangle_\phi$ with resepct to $x_0$ and show that this yields $x_0=0$.  Note that this is applying the variational principle to our trial wavefunction (to non-linear coefficient $x_0$).

   **$\frac{d\langle E \rangle_\phi}{dx_0} = \frac{d}{dx_0} \left(\frac{\hbar^2\alpha}{2\mu} + \frac{\mu\omega^2}{2}\left[\frac{1}{4\alpha} + x_0^2 \right] \right)\\
   = \mu\omega^2x_0$**
   
   **Setting this equal to zero yields**
   
   **$\frac{d\langle E \rangle_\phi}{dx_0} = 0 = \mu\omega^2x_0 \\
   \Rightarrow x_0 = 0$**
   
   **Since $\mu \neq 0$ and $\omega \neq 0$ (otherwise the potential is zero).**

4. Given all of the above, what is the variationally optimized wavefunction, $\phi(x)$?  Have we determined $c_0$?  If so, what is it?  If not, how can we determine it?  

   **Normalization constant.$c_0 = \left(\frac{2\alpha}{\pi}\right)^{1/4}= \left(\frac{\mu\omega}{\pi\hbar}\right)^{1/4}$ **

5. Compare optimized $\phi(x)$ and $\langle E \rangle_\phi$ to the ground-state harmonic oscillator solutions.

   **The trial wavefunction is the ground-state harmonic oscillator wavefunction if $x_0=0$ and $\alpha = \frac{\mu\omega}{2\hbar}$.**
