# A Particle in a Box

A particle restricted to be in a finite box is one the simplest problems to solve using quantum mechanical techniques.  It is thus a ubiquitous problem in introductory QM courses.  Here we will use it to solidfy some of the ideas discussed in the "Postulates and Formalism" class last Friday.  For the sake of this discussion we will restrict the particle to be in one dimension (1D) but in your homework you will do a problem with a particle in a 2D "box."

*Statement of the problem:* Consider a particle restricted to $0 \leq x \leq a$ in one dimension.  What are the energy levels of such a system?  What is the wave function for such a system? What is the average position and momentum of such a particle?  

## Solving the Schrodinger Equation

This can be thought of as a system with infinite potential (/repulsive) walls at $x=0$ and $x=a$ and zero potential otherwise.  Thus the Hamiltonian for such a system is simply the kinetic energy and the restriction of $0 \leq x \leq a$ will be applied to the bounds of the function.  Thus, 

$\hat{H} = -\frac{\hbar^2}{2m}\frac{d^2}{dx^2}$

Note that we only have one dimension, $x$, and thus the differential is written as $d$ instead of $\partial$ and there are no other components of $\nabla$.  Given this Hamiltonian, we can write the (time-independent) Schrodinger equation as:

$ -\frac{\hbar^2}{2m}\frac{d^2}{dx^2} \psi(x) = E \psi(x) \qquad 0 \leq x \leq a $

The goal now is to find a function, $\psi(x)$, that satisfies the above equation.  This means that it has to be an eigenfunction of the second derivative operator.  There are three obvious functions that return the same function upon taking the second derivative: $\sin$, $\cos$ and $\exp$.  Given that these functions are all possible solutions and that the second derivative is a linear operator, any linear combination of these three functions is also a possible solution to the above Schrodinger equation.  Luckily, these three functions are related by 

$e^{i\omega x} = \cos(\omega x) + i\sin(\omega x)$

and the second derivative of this function will return itself times a constant.  So the generic form of the solution to the wave function for a particle in a 1D box is

$\psi(x) = A\cos(\omega x) + B\sin(\omega x)$,

where $A$ and/or $B$ could be complex.  The eigenvalue problem dictates that $\omega = \frac{\sqrt{2mE}}{\hbar}$.  The boundaries give us additional restrictions on the wavefunction.  Namely, the wave function must be zero at the boundaries implying that $\psi(x=0) = 0$ and $\psi(x=a) = 0$.  The first of these restrictions implies that $A=0$ since $\cos(0) = 1$ and the second of which gives:


$B\sin(\omega a) = 0 \Rightarrow \omega a = n\pi \qquad n=1,2,3...$

Thus we get $\omega = \frac{n\pi}{a} = \frac{\sqrt{2mE}}{\hbar}$ or $E_n = \frac{h^2n^2}{8ma^2}$. Ultimately $\psi(x) = B\sin(\frac{n\pi}{a}x)$.  We determine the normalization constant, $B$, by enforcing that the square of the wave function integrate to 1,

$\int_0^a \psi(x)^*\psi(x) dx = B^2\int_0^a \sin^2(\frac{n\pi}{a} x)dx  = 1.$

We now let $z=\frac{n\pi x}{a}$ yielding

$ 1 = B^2\int_0^{n\pi} \sin^2(z x)dx = B^2\frac{a}{n\pi}\left(\frac{n\pi}{2}\right) = B^2\frac{a}{2}$

Or

$B = \sqrt{\frac{2}{a}}$.

So now we have:

$E_n = \frac{h^2n^2}{8ma^2}$ 

and

$\psi(x) = \sqrt{\frac{2}{a}}\sin(\frac{n\pi x}{a}) $.

This yields a probability distribution of: 

$\psi^2(x) = \frac{2}{a}\sin^2(\frac{n\pi x}{a}) $.

```python
# let's plot some probability distributions!
import matplotlib.pyplot as plt
import numpy as np

def psi2(x,n):
    return 2*np.sin(n*np.pi*x)**2

x = np.arange(0,1,0.01)
plt.plot(x,psi2(x,1),label="n=1")
plt.plot(x,psi2(x,2),label="n=2")
plt.plot(x,psi2(x,3),label="n=3")
plt.legend()
```

## Computing Expectation Values of Position and Momentum

***Average positions:*** Now let's compute the average position and position squared of the particle .  For this we need the position operator, $\hat{X} = x$.  We now compute the following:

$\langle x \rangle = \int_0^a \psi^*(x) \left[x\right]\psi(x) dx$

$ = \int_0^a  \sqrt{\frac{2}{a}}\sin(\frac{n\pi x}{a}) \cdot x \cdot \sqrt{\frac{2}{a}}\sin(\frac{n\pi x}{a}) dx$

$ = \frac{2}{a} \int_0^a x \sin^2(\frac{n\pi x}{a})dx $

$ = \frac{2}{a} \cdot \frac{a^2}{4}$

$ = \frac{a}{2} $

Similarly for average position squared, $\hat{X}^2 = x^2$ we get:

$\langle x^2 \rangle = \frac{2}{a} \int_0^a x^2 \sin^2(\frac{n\pi x}{a})dx $

$ = \left(\frac{a}{2\pi n}\right)\left(\frac{4\pi^2n^2}{3}-2\right) = \frac{a^2}{3} - \frac{a^2}{2n^2\pi^2}$

***Average momentum:*** Now let's compute the average linear momentum.  For this we need the linear momentum operator, $\hat{P}_x = -i\hbar\frac{d}{dx}$.  We now compute the following:

$\langle p \rangle = \int_0^a \psi^*(x) \left[-i\hbar\frac{d}{dx} \right]\psi(x) dx$

We note that since there is no imaginary component of $\psi(x)$ in this case that $\psi^*(x) = \psi(x)$.  Next we must compute the first derivative of $\psi(x)$:

$\frac{d}{dx} \psi(x) = \sqrt{\frac{2}{a}} \frac{d}{dx} \sin(\frac{n\pi x}{a}) = \sqrt{\frac{2}{a}} \frac{n\pi}{a} \cos(\frac{n\pi x}{a}).$

Thus, for the average momentum we get:

$\langle p \rangle = -i\hbar \frac{2n\pi}{a^2} \int_0^a \sin(\frac{n\pi x}{a})\cos(\frac{n\pi x}{a})dx  = 0$

How about determining $\langle p^2 \rangle$?  First we must recognize that $\hat{P}_x^2 = -\hbar^2\frac{d^2}{dx^2}$, then

$\langle p^2 \rangle = \int_0^a \psi^*(x) \left[-\hbar^2\frac{d^2}{dx^2} \right]^2\psi(x) dx$

$ = \hbar^2 \frac{2}{a} \left(\frac{n\pi}{a}\right)^2 \int_0^a \sin\frac{n\pi x}{a} \sin\frac{n\pi x}{a}dx $

$ =  \frac{2n^2\pi^2\hbar^2}{a^3}  \int_0^a \sin\frac{n\pi x}{a} \sin\frac{n\pi x}{a}dx $

$ = \frac{2n^2\pi^2\hbar^2}{a^3}\cdot \frac{a}{2} $

$ = \frac{n^2\pi^2\hbar^2}{a^2} $

## The Uncertainty Principle

The Heisenberg uncertainty principle can be stated mathematically as $\sigma_x\sigma_p > \hbar/2$.  In words, it can be stated the we cannot know the position and momentum of a quantum particle to infinite accuracy.  Indeed, as we become more sure of the position of the particle we must become less sure of the momentum.  The accuracy that we can know these two values is bounded by $\hbar/2$.  

$\sigma_x$ is the standard deviation of the position and $\sigma_p$ is the standard deviation of the momentum.  These can be defined as:

$\sigma_x^2 = \langle x^2 \rangle - \langle x \rangle^2$

$\sigma_p^2 = \langle p^2 \rangle - \langle p \rangle^2$

where $\sigma_x^2$ is called the variance of the position and is simply the standard deviation squared.  

We have computed the four avearge quantities of position, position squared, momentum and momentum squared for the particle in a 1D box.  Thus we can compute the product, $\sigma_x\sigma_p$, for this system and demonstrate that the Heisenberg inequality holds for it.  For the particle in a 1D box we have:

$\sigma_x^2 = \left[ \frac{a^2}{3} - \frac{a^2}{2n^2\pi^2} \right] - \left[ \frac{a}{2} \right]^2 $

$ = \left( \frac{a}{2n\pi} \right)^2 \left( \frac{\pi^2n^2}{3}-2 \right) $

$\sigma_p^2 = \left[ \frac{n^2\pi^2\hbar^2}{a^2} \right] - 0 $

$ = \frac{n^2\pi^2\hbar^2}{a^2} $

Thus for the product, $\sigma_x\sigma_p$, we get:

$\sigma_x\sigma_p = \sqrt{\left( \frac{a}{2n\pi} \right)^2 \left( \frac{\pi^2n^2}{3}-2 \right)}\sqrt{\frac{n^2\pi^2\hbar^2}{a^2}} $

$ = \frac{\hbar}{2}\left( \frac{\pi^2n^2}{3}-2 \right)^{1/2} $

We can see that the term inside the parantheses will always be greater than one for $n\geq1$, thus

$\sigma_x\sigma_p > \frac{\hbar}{2} $
